```python
import cv2
#import numpy.
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
bod1 = cv2.imread("board.png")
bod2 = cv2.imread("board2.png")
```

```python
bod1 = cv2.cvtColor(bod1,cv2.COLOR_BGR2GRAY)
bod2 = cv2.cvtColor(bod2,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(bod1,cmap="gray")
```

```python
plt.imshow(bod2,cmap="gray")
```

```python
thresh = 150 
max = 255
th, dst1 = cv2.threshold(bod1,thresh,max,cv2.THRESH_BINARY)
th, dst2 = cv2.threshold(bod2,thresh,max,cv2.THRESH_BINARY)
```

```python
plt.imshow(dst1,cmap="gray")
```

```python
plt.imshow(dst2,cmap="gray")
```

```python
dst = cv2.bitwise_xor(dst1,dst2)
```

```python
plt.imshow(dst,cmap="gray")
```
